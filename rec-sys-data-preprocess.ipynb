```python
import numpy as np
import pandas as pd
import os
import json
import csv
import re
from bs4 import BeautifulSoup
```

```python
def process_json_file(file_path):
    global author_id_map, writer
    with open(file_path, 'r') as f:
        data = json.load(f)
        if data.get('locale') != "en":
            return

        author_handle = data['authorHandle']
        if author_handle not in author_id_map:
            author_id_map[author_handle] = len(author_id_map) + 1

        title = BeautifulSoup(data['title'], 'html.parser').get_text()
        title = '"' + title.replace('"', "'") + '"'
        content = BeautifulSoup(data['content'], 'html.parser').get_text()
        content = re.sub(r'\$.*?\$', '', content)
        content = content.replace('\r\n', ' ').replace('\r', ' ').replace('\n', ' ')
        content = ' '.join(content.split())
        content = '"' + content.replace('"', "'") + '"'
        rating = data.get('rating', 0)
        tags = data.get('tags', [])

        writer.writerow({
            'id': data['id'],
            'title': title,
            'content': content,
            'authorHandle': author_handle,
            'authorID': author_id_map[author_handle],
            'rating': rating,
            'tags': tags
        })
```

```python
def process_files():
    total_files = 0
    for dirname, _, filenames in os.walk('/kaggle/input/rec-sys-blogs/data-final'):
        for filename in filenames:
            if filename.endswith('.json'):
                total_files += 1
                path = os.path.join(dirname, filename)
                print(f"\rFile being processing: {path}", end='', flush=True)
                process_json_file(path)
#                 print(f"\rFiles processed: {total_files}", end='', flush=True)
                     
```

```python
author_id_map = {}
with open('output.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['id', 'title', 'content', 'authorHandle', 'authorID', 'rating', 'tags']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    process_files()
```
