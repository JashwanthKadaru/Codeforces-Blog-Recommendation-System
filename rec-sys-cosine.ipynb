```python
import pandas as pd
import numpy as np
```

```python
def parse_array_string(s):
    array_str = s.strip('[]')
    array = np.array([float(x) for x in array_str.split()])
    return array
```

```python
df = pd.read_csv("/kaggle/input/recc-sys-vectorized/vectorized paras.csv")
```

```python
df['np_array'] = df['para_vector'].apply(parse_array_string)
print(df['np_array'].head)
```

```python
def cosine_similarity_custom(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    
    if norm_a == 0 or norm_b == 0:
        return 0  # Handle zero vector case to avoid division by zero
    
    cosine_sim = dot_product / (norm_a * norm_b)
    return cosine_sim
```

```python
def find_similar_elements(target_id, top_n=5):
    global df
    
    target_array = df.loc[df['id'] == target_id, 'np_array'].values[0]
    
    # Filter the DataFrame to exclude the target element
    df_filtered = df[df['id'] != target_id]

    # Calculate cosine similarity between the target array and all other arrays
    similarities = [cosine_similarity_custom(target_array, array) for array in df_filtered['np_array']]

    # Sort indices by similarity scores in descending order
    similar_indices = np.argsort(similarities)[::-1]

    # Get IDs and similarity scores of the top N most similar elements
    similar_ids = df_filtered.iloc[similar_indices[:top_n]]['id'].tolist()
    similar_scores = np.array(similarities)[similar_indices[:top_n]]

    return similar_ids, similar_scores
```

```python
target_id = 67654
top_n = 5

similar_ids, similarity_scores = find_similar_elements(target_id, top_n)

# Print content for target ID
print("Content for Target ID (", target_id, "):")
print(df.loc[df['id'] == target_id, 'content'].values[0])  # Assuming there's only one row for each ID

# Print content for similar IDs
for i, (id_, score) in enumerate(zip(similar_ids, similarity_scores), 1):
    print(f"\nSimilar element {i}: ID - {id_}, Similarity Score - {score}")
    print("Content:")
    print(df.loc[df['id'] == id_, 'content'].values[0])  # Assuming there's only one row for each ID
```
