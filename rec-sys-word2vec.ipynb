```python
import gensim
from gensim.models.doc2vec import TaggedDocument
import gensim.downloader as downloader
import numpy as np
import pandas as pd
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv("/kaggle/input/rec-sys-blogs/output.csv")
blogs = df['content']
# print(df.head)
```

```python
print(blogs.head)
```

```python
tagged_data = [TaggedDocument(words=gensim.utils.simple_preprocess(str(doc)), tags=[str(i)]) for i, doc in enumerate(blogs)]
```

```python
model = gensim.models.Doc2Vec(vector_size=100, window=5, min_count=2, epochs=40, workers=4)
model.build_vocab(tagged_data)
model.train(tagged_data, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
doc_vectors = [model.infer_vector(doc.words) for doc in tagged_data]
```

```python
df['para_vector'] = doc_vectors
```

```python
df.to_csv("vectorized paras.csv", index = False)
```
